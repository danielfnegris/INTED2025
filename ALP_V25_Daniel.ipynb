##estrura condicional

if - se

elif - senão se

else - senão

```python
nota = float(input("Digite sua nota: "))

if nota >= 7:
  print("Aprovado")
elif nota >= 5:
  print("Recuperação")
else:
  print("Reprovado")
```

4 - crie um algoritimo que receba uma idade do usuário, e depois verifique em qual faixa atária está à idade:

até 12 - você é um é uma criança

de 13 a 18 - você é um adolecente

de 19 á 60 - você é um adulto

acima de 60 - você é idoso

```python
idade = float(input("Digite sua idade: "))

if idade <= 12:
  print("Você é uma criança")
elif idade <=18:
  print("você é um adolescente")
elif idade <= 60:
  print("Você é um adulto")
else:
  print("Você é um idoso")
```
