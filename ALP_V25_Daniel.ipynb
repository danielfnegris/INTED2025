Peça a quantidade de horas trabalhadas e o valor por hora. Se trabalhou até 40 horas, calcule o salário normalmente (horas × valor por hora). Se trabalhou mais de 40 horas, as horas extras devem ter um adicional de 50% (horas extras × valor por hora × 1.5). Crie um algoritmo em Python que realize esse cálculo corretamente.

```python
NUM = int(input("Digite a quantidade de horas trabalhadas: "))
NUM2 = float(input("Digite o valor por hora: "))

if NUM <= 40:
  salario = NUM * NUM2
  print(f"Seu salário é de R${salario:.2f}")

elif NUM > 40:
  salario = (40 * NUM2) + ((NUM - 40) * (NUM2 * 1.5))
  print(f"Seu salário é de R${salario:.2f}")
```

##cire um algoritimo que faça uma calculadora, que repita até que o usuário digite 0 para sair

Deve mostrar o seguinte menu na tela:

Calculadora V25 digite:
1 - Multiplos de 5
2 - Número pares
3 - Números primos
4 - conversor de temperatura
0 - sair
--------------------------------------------
cada vez que o usuraio digitar um número deve entrar na função e pedir os dados ao usuário e efetuar o calculo.
mostrar a resposta e solicitar nvamnte o que ele quer


```python
x = -1

while x != 0:
    print("Calculadora V25 digite:")
    print("1 - Multiplos de 5")
    print("2 - Número pares")
    print("3 - Números primos")
    print("4 - conversor de temperatura")
    print("0 - sair")
    print("--------------------------------------------")

    x = int(input("Escolha uma opção: "))

    if x == 1:
        numero = int(input("Digite um número: "))
        print("Múltiplos de 5 até", numero)
        for i in range(0, numero + 1):
            if i % 5 == 0:
                print(i)
        print("--------------------------------------------")

    elif x == 2:
        numero = int(input("Digite um número: "))
        print("Números pares até", numero)
        for i in range(0, numero + 1):
            if i % 2 == 0:
                print(i)
        print("--------------------------------------------")

    elif x == 3:
        numero = int(input("Digite um número: "))
        print("Números primos até", numero)
        for num in range(2, numero + 1):
            primo = True
            for i in range(2, num):
                if num % i == 0:
                    primo = False
            if primo:
                print(num)
        print("--------------------------------------------")

    elif x == 4:
        print("1 - Celsius para Fahrenheit")
        print("2 - Fahrenheit para Celsius")
        tipo = int(input("Escolha a opção: "))

        if tipo == 1:
            c = int(input("Digite a temperatura em °C: "))
            f = (c * 9 // 5) + 32
            print(c, "°C equivale a", f, "°F")

        elif tipo == 2:
            f = int(input("Digite a temperatura em °F: "))
            c = (f - 32) * 5 // 9
            print(f, "°F equivale a", c, "°C")

        else:
            print("Opção inválida!")

        print("--------------------------------------------")

    elif x == 0:
        print("Saindo...")

    else:
        print("Opção inválida!")
        print("--------------------------------------------")
```
