##Estruta condicional

if  -  se

elif - senão se

else - senão

```python
nota = float(input("Digite sua nota: "))

if (nota >= 7):
  print("Aprovado")
elif (nota >= 5):
  print("Recuperação")
else:
  print("Reprovado")
```

4 - Crie um algoritmo que receba uma idade do usuário, e depois verifique em qual faixa etária está à idade:

até 12  - Você é uma criança!

de 13 à 18 - Você é um Adolescente!

de 19 à 60 - Você é um Adulto!

acima de 60 - Você é idoso!

```python
idade = int(input("Digite sua idade: "))

if (idade <= 12):
  print("Você é uma criança!")
elif (idade <= 18):
  print("Você é um Adolescente!")
elif (idade <= 60):
  print("Você é um Adulto!")
else:
  print("Você é idoso!")
```
